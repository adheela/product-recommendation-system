```python
# Import required libraries for text preprocessing
import re
import string
import pandas as pd

# NLTK for stopwords removal
import nltk
from nltk.corpus import stopwords

# Download stopwords (run once)
nltk.download('stopwords')

# Load English stopwords set
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    
    # 1. Convert text to lowercase
    text = text.lower()
    
    # 2. Remove punctuation and special characters
    text = re.sub(r'[^a-z\s]', '', text)
    
    # 3. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 4. Remove stopwords
    words = text.split()
    words = [word for word in words if word not in stop_words]
    
    return " ".join(words)
```

```python
df["text"] = (
    df["product_name"].astype(str) + " " +
    df["category"].astype(str) + " " +
    df["about_product"].astype(str)
)
```

```python
# Apply preprocessing to all rows in combined_text column

df["clean_text"] = df['review_content'].apply(clean_text)
```

3. Explain briefly why cosine similarity is used.

    Cosine similarity measures the angle between two vectors.

    ✔ It ignores magnitude (length of text)
    ✔ Focuses only on meaning similarity
    ✔ Works very well for TF-IDF text data
    ✔ Commonly used in recommendation systems

    Formula idea:
    cos(θ) = (A · B) / (||A|| ||B||)

```python
# -----------------------------
# Product Name Recommendation
# -----------------------------
def recommend(product_name, top_n=5):

    # fuzzy/partial match (important fix)
    matches = df[df["product_name"].str.contains(product_name, case=False, na=False)].index.tolist()

    if not matches:
        return "Product not found. Try a different name."

    idx = matches[0]

    # similarity scores
    sim_scores = list(enumerate(similarity_matrix[idx]))

    # sort by similarity (skip itself)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]

    recommended_idx = [i[0] for i in sim_scores]

    return df.loc[recommended_idx, ["product_name", "category", "rating"]]
```

```python
# -----------------------------

# TEST

# -----------------------------

print(recommend("boAt", 5))

print(recommend("Redmi", 5))

print(recommend("Samsung", 5))
```
